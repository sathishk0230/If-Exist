```python
import numpy as np
import random
import string
from collection import Collection
```

```python
l1_cnt = 10000000
l2_cnt = 10000
```

```python
l1 = [''.join(random.choices(string.ascii_uppercase + string.digits, k=20)) for i in range(l1_cnt)]
```

```python
l2 = list()
for _ in range(l2_cnt):
    if random.uniform(0,1) > 0.7:
        l2.append(random.choice(l1))
    else:
        l2.append(''.join(random.choices(string.ascii_uppercase + string.digits, k=20)))
```

```python
for l2ele in l2:
    if l2ele in l1:
        print("T",end='')
    else:
        print("F",end='')
```

### The above cell completed execution in 20min 5sec

```python
L1 = Collection(l1)
```

```python
for l2ele in l2:
    if l2ele in L1:
        print("T",end='')
    else:
        print("F",end='')
```
